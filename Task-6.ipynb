```python
#import all the libraries to read the data
import pandas as pd
```

```python
#load the data into pandas dataframe
df=pd.read_csv('Iris.csv')
```

```python
#print first five rows of data
df.head()
```

```python
#shape will return the total number of records available in the dataset and number of features
df.shape
```

```python
#let's check is there is any null value or not
df.isnull().sum()
```

```python
#there is no meaning of using id so simply drop it
df.drop(['Id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#to visualize the dataset
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(15,7))
ax1=fig.add_subplot(221)
ax1.hist(df['SepalLengthCm'],facecolor='peru',edgecolor='green',bins=10)
ax1.set_title('SepalLengthCm')
ax2=fig.add_subplot(222)
ax2.hist(df['SepalWidthCm'],facecolor='peru',edgecolor='green',bins=10)
ax2.set_title('SepalWidthCm')
ax3=fig.add_subplot(223)
ax3.hist(df['PetalLengthCm'],facecolor='peru',edgecolor='green',bins=10)
ax3.set_title('PetalLengthCm')
ax4=fig.add_subplot(224)
ax4.hist(df['PetalWidthCm'],facecolor='peru',edgecolor='green',bins=10)
ax4.set_title('PetalWidthCm')
```

```python
#create dependent and independent feature here we want to predict Species so it is our target variable(dependent)
x=df.drop(['Species'],axis=1)
y=df['Species']
```

```python
#to know what are different classes available in target variable
y.unique()
```

```python
#it is object format so we have to convert it into int. So here we are assigning different classes to every unique Species.
y.replace('Iris-setosa',0,regex=True,inplace=True)
y.replace('Iris-versicolor',1,regex=True,inplace=True)
y.replace('Iris-virginica',2,regex=True,inplace=True)
```

```python
y.unique()
```

```python
#split data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y , test_size=0.3) 
#now 30% of our data is test data
```

```python
#create DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
cls=DecisionTreeClassifier(random_state=42)
cls=cls.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
#plot the result of test data to see performance of algo on unseen data
plot_confusion_matrix(cls,x_test,y_test,values_format='d',display_labels=['Iris-setosa','Iris-versicolor','Iris-virginica'])
```

tree performace very good we can plot the tree now

```python
#plot the tree
from sklearn.tree import plot_tree
plt.figure(figsize=(25,20))
plot_tree(cls,filled=True,
          rounded=True,
          class_names=["Iris-setosa","Iris-versicolor","Iris-virginica"],
          feature_names=x.columns);
```

```python
#now we can check with any  value and predict the class
```
